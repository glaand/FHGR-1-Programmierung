## Aufgabe 16 - Schleifen

```python
def loopPower(base, numberOfLoops, search):
    alreadyFound = False
    for power in range(1, numberOfLoops):
        term = base**power
        rest = term % 1000
        print("Base: %d; Power: %d; Rest: %03d" % (base, power, rest))
        if power > base and rest == search and alreadyFound == False:
            alreadyFound = True
            print("Die Zahl '%d' ist die kleinste %der-Potenz die auf %03d endet." % (power,base,search))

numberOfLoops = 0
try:
    numberOfLoops = int(input("Anzahl Schleifendurchläufe: "))
except:
    raise Exception("Die eingegebene Zahl ist ungültig!")
        
loopPower(3, numberOfLoops, 1)
loopPower(7, numberOfLoops, 1)
```

## Aufgabe 17 - Zufallsstrings

```python
import random

lenStr = 0
possibleChars = ''

try:
    lenStr = int(input("Länge der zufällige Zeichenkette: "))
    possibleChars = input("Zeichenkette: ")
except:
    raise Exception("Fehler bei der Eingabe!")
    
generatedStr = ''
for i in range(lenStr):
    generatedStr += random.choice(possibleChars)
    
print(generatedStr)
```

## Aufgabe 18 - Verschachtelte for-Schleifen

```python
for i in range(1, 10 + 1):
    for j in range(1, 15 + 1):
        for k in range(1, 20 + 1):
            product = i*j*k
            squareRoot = product**(1/2)
            if squareRoot % 1 == 0:
                print("Quadratzahl: %d <- (i: %d, j: %d, k: %d)" % (product, i, j, k))
```

## Aufgabe 20 - (Schwierige Aufgabe) Häufigkeitstest eines Ereignisses

Ich bin mir nicht sicher, ob ich die negative Binomialverteilung richtig eingesetzt habe. Ohne den Location Parameter = 20, komme ich auf 100 für den Mittelwert, was angeblich im Vergleich zur Schleife nicht stimmt. Die Schleife tendiert zum Wert 120 als Mittelwert. Wahrscheinlich setzte ich die Verteilung falsch ein. Trotzdem habe ich was gelernt und merke das mir noch viel Wissen in der Stochastik fehlt.

Mit der Schleife habe ich folgendes gemacht. Ich habe 4 Listen erstellt, wo jede Liste Faktor 10 von der vorherigen Anzahl Stichproben entspricht. Das Resultat vom Code zeigt auf, dass desto höher die Anzahl Stichproben, desto eher tendiert der Mittelwert zur Zahl 120. Ich bin überzeugt, dass das Gesetz der grossen Zahlen hier eine wichtige Rolle spielt.

```python
# Anhand Negative Binomialverteilung
import numpy as np, scipy.stats as st

n = 20
p = 1/6

mean = st.nbinom.stats(n, p, moments='m', loc=20)
print("Durchschnittlich ungefähr %d-mal" % mean)
```

```python
# Anhand Schleife und Stichproben
import random

samples10 = []
samples100 = []
samples1000 = []
samples10000 = []

def getSample():
    matchedTimes = 0
    n = 1
    while True:
        if matchedTimes >= 20:
            break
        side = random.randint(1, 6)
        if side == 6:
            matchedTimes += 1
        n += 1
    return n

for i in range(10):
    samples10.append(getSample())
    
for i in range(100):
    samples100.append(getSample())
    
for i in range(1000):
    samples1000.append(getSample())
    
for i in range(10000):
    samples10000.append(getSample())
    
mean10 = sum(samples10) / len(samples10)
mean100 = sum(samples100) / len(samples100)
mean1000 = sum(samples1000) / len(samples1000)
mean10000 = sum(samples10000) / len(samples10000)

print("Durchschnittlich ungefähr %d-mal (Bei 10 Stichproben)" % mean10)
print("Durchschnittlich ungefähr %d-mal (Bei 100 Stichproben)" % mean100)
print("Durchschnittlich ungefähr %d-mal (Bei 1000 Stichproben)" % mean1000)
print("Durchschnittlich ungefähr %d-mal (Bei 10000 Stichproben)" % mean10000)
```
